## Preprocessing the dataset

```python
def tokenize(text):
    if not isinstance(text, str):
        return []
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()
    tokens = text.lower().translate(str.maketrans('', '', string.punctuation)).split()
    return list(set([w for w in tokens if w not in stop_words.STOP_WORDS and w.isalpha()]))

def preprocess(df):
    df.dropna(inplace=True)
    df["transaction"] = df["text"].apply(tokenize)
    transactions = df["transaction"].tolist()
    return transactions
```

```python
transactions = preprocess(df)
for i, t in enumerate(transactions):
    print(f"{i+1}: {t}")
```

## Analysis with PyFIM

```python
r = fpgrowth(transactions, supp=5)
df = pd.DataFrame(r)
df.columns = ['Itemset', 'Support']

df.sort_values(by='Support', ascending=False, inplace=True)
df
```

- Las palabras más frecuentes son "team" (equipo) y "ball" (balón), ya que son términos comunes en el fútbol.

- Un aspecto curioso es que la combinación de "away" y "team" (equipo visitante) es común, pero "home" (local) es más frecuente como un término aislado, lo que demuestra el poco uso del término "home team".

- Asimismo, se identifican palabras que representan elementos del juego como "keeper" (arquero), "game" (juego), "shot" (disparo), "crowd" (afición) y "midfield" (mediocampo).

- También resaltan términos utilizados para enfatizar las emociones, típico en comentarios deportivos: "truly" (verdaderamente) y "pure" (puro).

```python
r = fpgrowth(transactions, target='r', supp=2, report='aSc')
df = pd.DataFrame(r)
df.columns = ['Consequent', 'Antecedent', 'Freq', 'Freq(%)', 'Conf']
df.sort_values(by='Conf', ascending=False, inplace=True)
df
```

## Analysis with OR Tools

```python
rules = fpgrowth(transactions, target='r', supp=2, report='aSc')
rule_data = [
    {"ante": r[0], "cons": r[1], "support": r[2], "conf": r[3]}
    for r in rules if r[3] >= 0.5
]

solver = pywraplp.Solver.CreateSolver("SCIP")
x = [solver.BoolVar(f"x_{i}") for i in range(len(rule_data))]

solver.Maximize(solver.Sum(rule_data[i]["conf"] * x[i] for i in range(len(rule_data))))

solver.Add(solver.Sum(x) <= 100)

status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
    selected_rules = []
    for i in range(len(rule_data)):
        if x[i].solution_value() > 0:
            r = rule_data[i]
            selected_rules.append({
                "Consequent": r["cons"],
                "Antecedent": r["ante"],
                "Freq": r["support"],
                "Freq %": r["support"] * 100 / len(transactions),
            })
    selected_df = pd.DataFrame(selected_rules)
    display(selected_df)
else:
    print("No optimal solution found.")
```
